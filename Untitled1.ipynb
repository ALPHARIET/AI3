```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/AI/AI3/insurance.csv')

x = df['bmi'].values
y = df['charges'].values
```

```python
import numpy as np

w = 0
b = 0
lr = 0.0001
epochs = 1000

n = len(x)

for epoch in range(epochs):
    for i in range(n):
        y_pred = w * x[i] + b
        error = y_pred - y[i]

        dw = 2 * error * x[i]
        db = 2 * error

        w = w - lr * dw
        b = b - lr * db

print("Final weight:", w)
print("Final bias:", b)
```

```python
import matplotlib.pyplot as plt

y_pred = w * x + b

plt.scatter(x, y, label='Data Asli')
plt.plot(x, y_pred, color='red', label='Regresi Linear')
plt.xlabel('BMI')
plt.ylabel('Charges')
plt.legend()
plt.show()
```
